# Feature engineering

```python
# Renaming Columns
df =df.rename({'YearsExperience':'Years_Experience','Salary':'Salary_hike'},axis=1)
df.head()
```

# Original Data Visualization

```python
sns.distplot(df['Years_Experience'], bins = 10, kde = True)
plt.title('Before Transformation')
```

```python
sns.distplot(df['Salary_hike'], bins = 10, kde = True)
plt.title('Before Transformation')
```

```python
from sklearn import linear_model
reg = linear_model.LinearRegression()
```

```python
#creating & traning linear regrisson model
reg.fit(df[['Years_Experience']],df.Salary_hike)
```

```python
#prediction using linear regrisson model
#lets predict salary for 20 year of ecperience
reg.predict([[20]])
```
